```python
#df_sample.isna().sum()
df_sample_small = df_sample[['FIPS_txt','White %']]
df_sample_small = df_sample_small.dropna()
df_sample_small.isna().sum()
df_sample_small.head()
```

```python
colorscale = ["#f7fbff", "#ebf3fb", "#deebf7", "#d2e3f3", "#c6dbef", "#b3d2e9", "#9ecae1",
    "#85bcdb", "#6baed6", "#57a0ce", "#4292c6", "#3082be", "#2171b5", "#1361a9",
    "#08519c", "#0b4083", "#08306b"
]
endpts = list(np.linspace(1, 12, len(colorscale) - 1))
fips = df_sample_small['FIPS_txt'].tolist()
values = df_sample_small['White %'].tolist()

fig = ff.create_choropleth(
    fips = fips, values = values, scope = ['usa'],
    binning_endpoints = endpts, colorscale = colorscale,
    show_state_data = False,
    show_hover = True, centroid_marker = {
        'opacity': 0
    },
    asp = 2.9,
    title = 'SVB Race Stuff',
    legend_title = 'Race % White'
)
py.iplot(fig, filename = 'choropleth_full_race')
```
